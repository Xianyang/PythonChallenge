There is a strange picture with a gray bar on the web page. And there is basically nothing in the page soucre. So the answer is in the image. I use [PCL(Python Imaging Library](http://www.pythonware.com/products/pil/) to read this image. Read the gray bar and output the list as below

### Solution 1

```python
from PIL import Image
import re, StringIO, urllib
im = Image.open('../Resource/oxygen.png')

bar = [im.getpixel((i, 47)) for i in range(im.size[0])]
print bar
```

It's easy to find out that the gray pixels have R=G=B and all the piexls have a period of 7. So use a loop to get the numbers and transfer the number to char. A message will show. 

```python
line = [im.getpixel((i, 47)) for i in range(0, im.size[0], 7)]
ords = [r for r, g, b, a in line if r == g == b]
message = ''.join(map(chr, ords))
print message
```

Use the same way to transfer the numbers above.

```python
number_in_message = re.findall('\d+', message)
final_message = map(int, number_in_message)
print ''.join(map(chr, final_message))
```
